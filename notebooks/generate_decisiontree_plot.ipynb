```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mlxtend.plotting import plot_decision_regions
```

```python
df_wine = pd.read_csv('../data/wine_dataset.csv',usecols=['alcohol','ash','proline','hue','class'])

X = df_wine[['proline','hue']]
y = df_wine['class']

y.value_counts().sort_index()
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X,y)

fig,ax = plt.subplots(1,1,figsize=(6,6))
plot_decision_regions(X.values, y.values, clf=dtc);
plt.xlabel(X.columns[0]); plt.ylabel(X.columns[1]);
```

```python
from sklearn.tree import plot_tree

fig,ax = plt.subplots(1,1,figsize=(20,12))
plot_tree(dtc,feature_names=X.columns,filled=True,fontsize=10,ax=ax,);

fig.savefig('images/dt_wine_plot_maxdepth10.png')
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=3)
dtc.fit(X,y)

fig,ax = plt.subplots(1,1,figsize=(6,6))
plot_decision_regions(X.values, y.values, clf=dtc);
plt.xlabel(X.columns[0]); plt.ylabel(X.columns[1]);
```

```python
from sklearn.tree import plot_tree

fig,ax = plt.subplots(1,1,figsize=(20,12))
plot_tree(dtc,feature_names=X.columns,filled=True,fontsize=18,ax=ax,);

fig.savefig('images/dt_wine_plot_maxdepth3.png')
```
